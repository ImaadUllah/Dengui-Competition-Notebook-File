```python
train_data.fillna(method='ffill', inplace=True)

test_data.fillna(method='ffill', inplace=True)
```

```python
train_data['city'] = train_data['city'].astype('category') 
train_data['city'] = train_data['city'].cat.codes

test_data['city'] = test_data['city'].astype('category') 
test_data['city'] = test_data['city'].cat.codes


```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_training = train_data.drop(['total_cases'], 1)
y_training = train_data['total_cases']
x_train, x_test, y_train, y_test = train_test_split(x_training, y_training, test_size=0.2852, random_state=1)
```

```python
from sklearn.metrics import mean_absolute_error
regression_model = LinearRegression()
regression_model.fit(x_train, y_train)
y_predicted = regression_model.predict(x_test)

print('Mean Absolute Error:', mean_absolute_error(y_test, y_predicted))
```

```python
regression_model.fit(x_training, y_training)
y_predicted = regression_model.predict(test_data)

print('Mean Absolute Error:', mean_absolute_error(y_test, y_predicted))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
clf = linear_model.Ridge(alpha = 44)
clf.fit(x_train, y_train)
Y_pred = clf.predict(x_test)
mean_absolute_error(y_test, Y_pred)
```

```python
clf.fit(x_training, y_training)
Y_pred = clf.predict(test_data)
mean_absolute_error(y_test, Y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor(max_depth=2, random_state=0, n_estimators=100)
RFR.fit(x_train, y_train)
y_ = RFR.predict(x_test)
print('Mean Absolute Error:', mean_absolute_error(y_test, y_))
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor(max_depth=2, random_state=42, n_estimators=100)
RFR.fit(x_training, y_training)
y_ = RFR.predict(test_data)
print('Mean Absolute Error:', mean_absolute_error(y_test, y_))
```

```python
from sklearn.model_selection import GridSearchCV
RFclfParameters = [{'n_estimators': [100, 500], 'max_features': ["log2","sqrt"]}]
GridSearchRF = GridSearchCV(estimator = RFR, param_grid = RFclfParameters, scoring = 'neg_mean_absolute_error', cv = 10, n_jobs = -1)
```

```python
GridSearchRF.fit(x_train, y_train)
```

```python
yy = GridSearchRF.predict(x_test)
print('Mean Absolute Error:', mean_absolute_error(y_test, yy))
```

```python
GridSearchRF.fit(x_training, y_training)
y = GridSearchRF.predict(test_data)
print('Mean Absolute Error:', mean_absolute_error(y_test, y))
```
